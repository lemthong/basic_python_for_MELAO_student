# 1. Energy generation in Laos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Energy generation_LAO.csv', on_bad_lines='skip')
data
```

## 2. Change columns name

```python
data = data.rename(columns={"EDL and EDL+Gen":"EDL_Gen"})
data
```

## 3. Plot bar graph

```python
fig, ax = plt.subplots(figsize=(10, 5))
years = data['Period']
x = np.arange(len(years))
width = 0.3
ax.bar(x - 3*width/2, data['Total'], width, label='Total', color='#0343df')
ax.bar(x - width/2, data['EDL_Gen'], width, label='EDL_Gen', color='#e50000')
ax.bar(x + width/2, data['IPP'], width, label='IPP (Independent Power Producer)', color='#ffff14')

ax.set_ylabel('Giga Watt')
ax.set_title('Energy generation in Laos')

ax.set_xticks(x)    # This ensures we have one tick per year, otherwise we get fewer
ax.set_xticklabels(years.astype(str).values, rotation='vertical')
ax.grid(color='#cccccc')
ax.legend()



plt.show()
```

## 4. Sharing rate of power generation

```python
data['% EDL_Gen'] = data['EDL_Gen'] * 100 / data['Total']
data
```

```python
data['% IPP'] = data['IPP'] * 100 / data['Total']
data
```

```python
data1=data.drop(columns =['EDL_Gen', 'IPP'])
data1
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
years = data['Period']
x = np.arange(len(years))
width = 0.3

ax.bar(x - width/2, data['% EDL_Gen'], width, label='% EDL_Gen', color='#e50000')
ax.bar(x + width/2, data['% IPP'], width, label='% IPP', color='#ffff14')

ax.set_xlabel('Year', size=13)
ax.set_ylabel('%', size=13)
ax.set_title('Shared energy generation in Laos', size=18)

ax.set_xticks(x)    # This ensures we have one tick per year, otherwise we get fewer
ax.set_xticklabels(years.astype(str).values, rotation='vertical')
ax.grid(color='#cccccc')
ax.legend()

plt.show()
```
